```python
import pandas as pd
import seaborn as sns
```

```python
coocc_pop = pd.read_csv('../results/coocc/coocc_Pop.csv', index_col=0)
coocc_rock = pd.read_csv('../results/coocc/coocc_Rock.csv', index_col=0)
coocc_country = pd.read_csv('../results/coocc/coocc_Country.csv', index_col=0)
coocc_electronic = pd.read_csv('../results/coocc/coocc_Electronic.csv', index_col=0)
```

```python
all_cols = list(set([*coocc_pop.columns, *coocc_rock.columns, *coocc_electronic.columns, *coocc_country]))
```

```python
prob_rock = coocc_rock.reindex(columns=all_cols, fill_value=0)
prob_rock = prob_rock.T
prob_rock = prob_rock.reindex(columns=all_cols, fill_value=0)
prob_rock = prob_rock.T
```

```python
prob_pop = coocc_pop.reindex(columns=all_cols, fill_value=0)
prob_pop = prob_pop.T
prob_pop = prob_pop.reindex(columns=all_cols, fill_value=0)
prob_pop = prob_pop.T
```

```python
prob_country = coocc_country.reindex(columns=all_cols, fill_value=0)
prob_country = prob_country.T
prob_country = prob_country.reindex(columns=all_cols, fill_value=0)
prob_country = prob_country.T
```

```python
prob_electronic = coocc_electronic.reindex(columns=all_cols, fill_value=0)
prob_electronic = prob_electronic.T
prob_electronic = prob_electronic.reindex(columns=all_cols, fill_value=0)
prob_electronic = prob_electronic.T
```

```python
prob_rock
```

# Rock - Pop

```python
diff = prob_rock - prob_pop
```

```python
diff
```

```python
mask = ((diff > 0.01) | (diff < -0.01))
diff_filtered = diff[mask.any()].T[mask.any()].T
```

```python
_ = sns.heatmap(diff_filtered, cmap='Greys', cbar_kws={'label': 'Difference in occurence in Rock vs. Pop (%)'}, annot=True)
```

# Rock - Country

```python
diff_pop_rock = prob_rock - prob_country
```

```python
mask = ((diff > 0.01) | (diff < -0.01))
diff_filtered = diff[mask.any()].T[mask.any()].T
```

```python
_ = sns.heatmap(diff_filtered, cmap='Greys', annot=True)
```

This means that e.g. the transition IV - I happens more often percentage-wise in rock than in country music.

# Country - Pop

```python
diff = prob_country - prob_pop
```

```python
mask = ((diff > 0.03) | (diff < -0.015))
diff_filtered = diff[mask.any()].T[mask.any()].T
```
